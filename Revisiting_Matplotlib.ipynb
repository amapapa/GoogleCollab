```
## Now that we have some test data, lets try plotting them in some format

x=np.linspace(0,2,100)
print(x)

plt.plot(x)
```

```

plt.plot(x,x,'r',label='linear')
plt.plot(x, x**2, 'b',label='quadratic')
plt.plot(x, x**3,'c', label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

plt.legend()



plt.show()
```

```
x=np.arange(0,10,0.2)
# print(x)
y=np.sin(x)
print(y)

plt.plot(x,y,'g-')
plt.xlabel("X Label")
plt.ylabel("Y Label")

plt.title("Sine wave")

plt.show()
```
